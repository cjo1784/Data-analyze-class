#3. 그래프에 옵션 추가하기
1. 그래프에 제목 넣기
 - title()함수를 사용하면 그래프에 제목을 추가할 수 있음.

```python
import matplotlib.pyplot as plt
plt.title("graph")
plt.plot([10,20,30,40])
plt.show()
```

2. 그래프에 범례 넣기
 - 여러 그래프를 구분하기 위해서 범례(legend)를 추가함
 - 범례란 사용자가 인지하기 쉽도록 본보기로 표시하는데 사용하는것을 말한다.
 - plot()함수에 label이라는 속성의 레이블 값으로 원하는 문자열을 넣어줌
 - 그래프를 그리기 전에 legend()함수를 실행시키면 레이블 값이 범례로 나타남.


```python
import matplotlib.pyplot as plt
plt.title("legend")
plt.plot([10,20,30,40],label="acs")
plt.plot([40,30,20,10],label="decs")
plt.legend()
plt.show()
```

3. 그래프 색상 바꾸기
 - color 속성을 추가하면 선의 색을 변경할 수 있음.
 - color 속성을 생략하면 자동으로 색상이 설정됨.



```python
import matplotlib.pyplot as plt
plt.title("color")
plt.plot([10,20,30,40],color="skyblue",label="skyblue")
plt.plot([40,30,20,10],'pink',label="pink")
plt.legend()
plt.show()
```

4. 그래프 선 모양 바꾸기
 - linestyle 속성에 원하는 선 모양을 지정하면 선 모양을 바꿀 수 있음.
 - linestyle 대신 ls라고 작성할 수 있음.



```python
import matplotlib.pyplot as plt
plt.title("linestyle")
plt.plot([10,20,30,40],color="r",linestyle='--',label="dashed")
plt.plot([40,30,20,10],color='g',ls=':',label="dotted")
plt.legend()
plt.show()
```

5. 마커 모양 바꾸기
 - marker속성을 설정하면 선 형태가 아닌 점 형태로 그래프 표현이 가능
 - 마커 모양으로는 선, 점, 동그라미, 삼각형, 별 모양 등 여러가지가 존재



```python
import matplotlib.pyplot as plt
plt.title("marker")
plt.plot([10,20,30,40],"r",marker=".",label="circle") # 빨간색 원형 마커 그래프
plt.plot([40,30,20,10],'g',marker="^",label="triangle up") # 초록색 삼각형 마커 그래프
plt.legend()
plt.show()
```

- 특정 날짜를 입력하여서 해당 날짜의 데이터들을 리스트에 담아 그래프로 나타내보자.

```python
import csv
import matplotlib.pyplot as plt
f = open("/content/seoul_temp.csv.csv",'r',encoding="cp949")
data = csv.reader(f)
next(data)
result = []
mm,dd = input().split()
for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == mm and row[0].split('-')[2] == dd:
      result.append(float(row[-1]))
plt.figure(figsize=(10, 6))
plt.plot(result,'r')
plt.show()
```

- 년도의 데이터도 너무 많다. 특정 년도를 정하여 해당 년도부터 현재까지 특정한 날짜의 최고, 최저 기온을 년도별 기준을 가지고 그래프로 그려보자.

```python
import csv
import matplotlib.pyplot as plt
f = open("/content/seoul_temp.csv.csv",'r',encoding="cp949")
data = csv.reader(f)
next(data)
high = []
low = []
yyyy,mm,dd = input().split()
for row in data:
  if row[-1] != '':
    if int(yyyy) <= int(row[0].split('-')[0]):
      if row[0].split('-')[1] == mm and row[0].split('-')[2] == dd:
        high.append(float(row[4]))
        low.append(float(row[3]))
plt.plot(high,'hotpink',label="high")
plt.plot(low,'skyblue',label="skyblue")
plt.legend()
plt.show()
```

- 동그라미의 의미는 이상치(outlier)값을 표현한 것으로, 다른 수치에 비해 너무 크거나 너무 작은 값을 자동으로 나타낸 것
- boxplot을 따로 하는게 아니라 하나의 함수로 2개(1월,8월)를 묶어보자

```python
import csv
import matplotlib.pyplot as plt

f = open("seoul_temp.csv",'r',encoding="cp949")
data = csv.reader(f)
next(data)
aug = []
jan = []
for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '08':
      aug.append(float(row[-1]))
    if row[0].split('-')[1] == '01':
      jan.append(float(row[-1]))
plt.boxplot([jan,aug])
plt.show()
```

- 최고의 기온 데이터를 월별로 구분하여 보자 다음 2가지 조건을 만족해야 한다

  1) 데이터를 월별로 분류해 저장한다.

  2) 월별 데이터를 상자 그림으로 그린다.

```python
import matplotlib.pyplot as plt
import csv

f = open("seoul_temp.csv",'r',encoding="cp949")
data = csv.reader(f)
next(data)

month = [[],[],[],[],[],[],[],[],[],[],[],[]] #각 12개의 월을 입력받을 빈 리스트 생성

for row in data:
  if row[-1] != '':
    month[int(row[0].split('-')[1])-1].append(float(row[-1]))
plt.boxplot(month)
plt.show()
```
